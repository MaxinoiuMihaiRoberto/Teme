```python
import pandas as pd
import numpy as np
```

```python
drinks=pd.read_csv('drinks.csv', sep='\s*,\s*')
print(drinks)
```

```python
df=pd.DataFrame(drinks)
print(df)
```

```python
ContinentEU = df[df['continent']=='EU']
ContinentAS = df[df['continent']=='AS']
ContinentAF = df[df['continent']=='AF']
ContinentSA = df[df['continent']=='SA']
ContinentOC = df[df['continent']=='OC']
ContinentNA = df[df['continent']=='NA']


print(ContinentOC)

```

```python
AverageBeerEU=ContinentEU['beer_servings'].sum()

AverageBeerAS=ContinentAS['beer_servings'].sum()

AverageBeerAF=ContinentAF['beer_servings'].sum()

AverageBeerSA=ContinentSA['beer_servings'].sum()

AverageBeerOC=ContinentOC['beer_servings'].sum()

AverageBeerNA=ContinentNA['beer_servings'].sum()
```

```python
A={'Continent':['EU','AS','AF','SA','OC','NA'],
   'AverageBeerPerContinent':[AverageBeerEU,AverageBeerAS,AverageBeerAF,AverageBeerSA,AverageBeerOC,AverageBeerNA]}
AverageBeer=pd.DataFrame(A)
print(AverageBeer)
```

```python
print(AverageBeer['AverageBeerPerContinent'].max())
```

```python
AverageWineEU=ContinentEU['wine_servings'].sum()

AverageWineAS=ContinentAS['wine_servings'].sum()

AverageWineAF=ContinentAF['wine_servings'].sum()

AverageWineSA=ContinentSA['wine_servings'].sum()

AverageWineOC=ContinentOC['wine_servings'].sum()

AverageWineNA=ContinentNA['wine_servings'].sum()
```

```python
B={'Continent':['EU','AS','AF','SA','OC','NA'],
   'AverageWinePerContinent':[AverageWineEU,AverageWineAS,AverageWineAF,AverageWineSA,AverageWineOC,AverageWineNA]}
AverageWine=pd.DataFrame(B)
print(AverageWine)
```
